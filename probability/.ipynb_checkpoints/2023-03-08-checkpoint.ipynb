# 二项分布和泊松分布的近似效果

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import stats
```

由上课老师讲的可以知道 n越大 p越小，拟合效果就越好
> 当n=20 ，p = 0.3 时

```python
n = 20
p = 0.3
bino = stats.binom(n,p)
x = np.arange(0,n)
y_bino = bino.pmf(x)
plt.plot(y_bino,label = '二项分布')
pois = stats.poisson(n*p)
y_pois = pois.pmf(x)
plt.plot(y_pois,label = '泊松分布')
plt.title("n=20 , p=0.3")
```

## 可以看到并没有很拟合
> n=100,p=0.05

```python
n = 100
p = 0.05
bino = stats.binom(n,p)
x = np.arange(0,n)
y_bino = bino.pmf(x)
plt.plot(y_bino,label = '二项分布')
pois = stats.poisson(n*p)
y_pois = pois.pmf(x)
plt.plot(y_pois,label = '泊松分布')
plt.title("n=100 , p=0.05")
```

这样就相当拟合了
